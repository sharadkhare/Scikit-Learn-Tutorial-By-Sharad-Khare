```python
# Business Problem Statement:
'''Let's consider a scenario where a company wants to understand 
the underlying structure of its customer data for targeted marketing.
The business problem is to reduce the dimensionality of the customer 
data while preserving the local relationships between customers.'''
# Locally Linear Embedding (LLE)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import LocallyLinearEmbedding
```

```python
# generating datasets
X,color= make_swiss_roll(n_samples=1000,random_state=42)
```

```python
lle = LocallyLinearEmbedding(n_neighbors=10, n_components=2, random_state=42)
X_lle = lle.fit_transform(X)
```

```python
# visualization
fig,(ax1,ax2) = plt.subplots(1,2, figsize=(12,5))
ax1.scatter(X[:,0],X[:,2],c=color,cmap=plt.cm.Spectral)
ax2.scatter(X_lle[:,0],X_lle[:,1],c=color,cmap=plt.cm.Spectral)
ax2.set_title('Reduced Dimensionaly using LLE')
plt.show()
```

```python
# user function for input
def project_new_customer(lle_model,new_customer_data):
    reduced_data = lle_model.transform(new_customer_data.reshape(1,-1))
    return reduced_data

new_customer = np.array([10.0,20.0,30.0])
reduced_new_customer= project_new_customer(lle,new_customer)
print("Projected data for new customer:")
print(reduced_new_customer)
```
